```python
!pip install streamlit
import streamlit as st
import pandas as pd

# Replace 'your_username' with your actual username
file_path ='/Users/florian_pothin/Desktop/df_for etu.xlsx'

# Read the Excel file
df = pd.read_excel(file_path)

# Title of the Streamlit app
st.title('Exploratory Data Analysis of DataFrame')

# Display the dataframe
st.write('DataFrame:')
st.dataframe(df)

# Display basic statistics
st.write('Basic Statistics:')
st.write(df.describe())

# Display column names
st.write('Column Names:')
st.write(df.columns)

# Display data types
st.write('Data Types:')
st.write(df.dtypes)

# Display missing values
st.write('Missing Values:')
st.write(df.isnull().sum())

# Filter data by column
column = st.selectbox('Select a column to filter by:', df.columns)
unique_values = df[column].unique()
selected_value = st.selectbox('Select a value:', unique_values)
filtered_df = df[df[column] == selected_value]
st.write(f'Data filtered by {column} = {selected_value}:')
st.dataframe(filtered_df)
```
